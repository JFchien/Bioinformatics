## count matching patterns

```python
def PatternCount(Text, Pattern):
    count=0
    pos=list()
    a=len(Text)
    b=len(Pattern)
    c=a-b
    for i in range(c+1):
        if Text[i:i+len(Pattern)]== Pattern:
            pos.append(str(i))
            count =count + 1
    return count,pos
```

## count frequency

```python
## O(n^2)
def FrequentWords(Text, k):
    FrequentPatterns = list()
    Count=list()
    for i in range(len(Text)-k):
        Pattern=Text[i:i+k]
        Count.append(PatternCount(Text, Pattern))
    maxCount=max(Count)
    for i in range(len(Text)-k):
        if (Count[i] == maxCount):
            FrequentPatterns.append(Text[i:i+k])
    return FrequentPatterns
```

```python
def Countkmer(Text,k):
    l=list()
    for i in range(len(Text)):
        l.append(Text[i:i+k])
    return pd.DataFrame(l).value_counts().head()
Countkmer('CGGAGGACTCTAGGTAACGCTTATCAGGTCCATAGGACATTCA',3)
```

```python
def FrequencyTable(Text, k):
    freqMap={}
    n= len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        if not freqMap.get(Pattern):
            freqMap[Pattern]=1
        else:
            freqMap[Pattern]=freqMap[Pattern]+1 
#             freqMap=dict(sorted(freqMap.items(), key=lambda freqMap: freqMap[1],reverse=True))
    return freqMap
# FrequencyTable('CGGACTCGACAGATGTGAAGAACGACAATGTGAAGACTCGACACGACAGAGTGAAGAGAAGAGGAAACATTGTAA',5)
```

```python
def FindClumps(Text, k, L, t):
    Patterns=list()
    n = len(Text)
    for i in range(n-L):
        Window = Text[i: i+L]
        freqMap = FrequencyTable(Window, k)
        for s in freqMap.keys():
            if (freqMap[s] >= t) & (s not in Patterns):
                Patterns.append(s)
    return Patterns
text_file = open("C:/Users/Jofan/Downloads/E_coli.txt", "r")
Text = text_file.read()  #read whole file to a string
text_file.close() #close file
FindClumps(Text,9,500,3)
```

```python
def ReverseComplement(Text):
    compli_dic={'A':'T','T':'A','C':'G','G':'C'}
    reverse_complement = "".join(compli_dic.get(base) for base in reversed(Text))
    return reverse_complement
```
